Task:

Create a dictionary with keys representing brand names.

For each brand key, store a list of 9 values.

Each value should be a string in the format:

nameX_YYYY

where:

X is a number (e.g., 7, 12, etc.).

YYYY is a year (e.g., 2019, 2020).

Use multiple brand names as dictionary keys (e.g., "Nike", "Adidas", "Puma").

Convert the dictionary into a Pandas DataFrame.


---




```python
import pandas as pd
dict_a = {'brand' : ['Lamborghini_1998', 'Porsche_2014', 'Volkswagen_2015', 'Chrysler_2009', 'volvo_1927']}
df = pd.DataFrame(dict_a)
df
```

```python
df[['brand_name', 'year']] = df['brand'].str.split("_", expand=True)
df
```

```python
df['brand_name_only'] = df['brand'].str.split("_").str[0]
df['year_only'] = df['brand'].str.split("_").str[1].astype("int")
df
```

```python
import pandas as pd

# Create date range
days = pd.date_range(start="2025-01-01", end="2025-01-31")

# Make DataFrame
df = pd.DataFrame({'Date': days.astype(str)})

# Split 'Date' column into Year, Month, Day
df[['Year', 'Month', 'Day']] = df['Date'].str.split('-', expand=True)

df
```
