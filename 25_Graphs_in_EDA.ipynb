# Graphs in Exploratory Data Analysis (EDA)
This notebook demonstrates how various graphs are used in data analysis using the Housing Prices dataset. It includes visualizations to explore relationships, distributions, comparisons, and category compositions.


---



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

house_data = pd.read_csv('/content/Housing.csv')
house_data.head()
```



---


### 📊 Graphs by Purpose
---



### 1. 🔁 Relationship (Regression / Prediction)


---



```python
sns.scatterplot(x='area', y='price', data=house_data)
house_data.plot(x='area', y='price', kind='scatter')
plt.show()
```

```python
sns.scatterplot(x='furnishingstatus', y='price', data=house_data)
plt.show()
```



---


### 2. 📈 Trend (Line Plot)


---



```python
house_data['price'].plot(kind='line')
```

```python
house_data['area'].plot()
```



---
### 3. 📊 Distribution (Histogram / KDE)


---




```python
house_data['price'].plot(kind='hist', bins=10)
sns.histplot(house_data['price'], kde=True)
sns.kdeplot(house_data['price'])
```

```python
house_data['price'].plot(kind='hist', bins=10)
```

```python
sns.histplot(house_data['price'],kde=True)
```

```python
sns.kdeplot(house_data['price'])
```



---
### 4. 📦 Category Comparison (Bar/Count/Box/Violin)


---




```python
# Bar
house_data['airconditioning'].value_counts().plot(kind='bar')
```

```python
# Count
sns.countplot(x='furnishingstatus', data=house_data)
```

```python
# Box
sns.boxplot(x='furnishingstatus', y='price', data=house_data)
```

```python
# Violin
sns.violinplot(x='furnishingstatus', y='price', data=house_data)
```



---

### 5. 📚 Multiple Relationships (Pair Plot)


---



```python
sns.pairplot(house_data)
```

### Summary Table


---



| **Graph Type**   | **Code Example**                                                        | **Use Case / Purpose**                                        | **Applied On**                         | **ML / Stats Relevance**                                      |
| ---------------- | ----------------------------------------------------------------------- | ------------------------------------------------------------- | -------------------------------------- | ------------------------------------------------------------- |
| **Scatter Plot** | `sns.scatterplot(x='area', y='price')`<br>`df.plot(kind='scatter')`     | Shows relationship between two **continuous** variables       | Continuous data (e.g., area vs. price) | Check **correlation**, used in **regression**                 |
| **Line Plot**    | `df['price'].plot(kind='line')`                                         | Trend over time or index                                      | Continuous (often time series)         | Useful for **time-based predictions**                         |
| **Histogram**    | `df['price'].plot(kind='hist', bins=10)`<br>`sns.histplot(df['price'])` | Distribution of a single variable                             | Continuous or Discrete                 | Understanding **probability distribution**                    |
| **Box Plot**     | `sns.boxplot(x='furnishingstatus', y='price')`                          | Shows median, quartiles, outliers                             | Mix of Categorical + Continuous        | Detect **outliers**, important in **data cleaning**           |
| **Count Plot**   | `sns.countplot(x='furnishingstatus')`                                   | Frequency of each category                                    | Categorical                            | Useful for **class imbalance** in classification              |
| **Bar Plot**     | `df['airconditioning'].value_counts().plot(kind='bar')`                 | Visualize category counts or aggregated values                | Categorical                            | Common in **categorical analysis**                            |
| **Violin Plot**  | `sns.violinplot(x='furnishingstatus', y='price')`                       | Combines boxplot + KDE to show distribution                   | Mix of Categorical + Continuous        | Compare **distribution across categories**                    |
| **KDE Plot**     | `sns.kdeplot(df['price'])`                                              | Smooth version of histogram; shows **probability density**    | Continuous                             | Highlights **distribution shape**, relates to **probability** |
| **Pair Plot**    | `sns.pairplot(df)`                                                      | Multiple scatter plots + histograms for feature relationships | Mixed Data                             | Quick way to inspect **correlation**, outliers                |
| **Bar Chart**    | Same as countplot, often `plot(kind='bar')`                             | Compare numerical values across categories                    | Categorical                            | Supports **comparison-based** ML decisions                    |

### 💡 How This Links to ML Concepts


---



| **Concept**                 | **Graph Example or Related Task**                                                |
| ------------------------------------ | -------------------------------------------------------------------------------- |
| **Probability**                      | Histogram, KDE, Violin, all show how data is distributed (basis of probability) |
| **Classification**                   | Count Plot, Bar Plot, understand class balance, a key in classification         |
| **Model Evaluation (Cross-Entropy)** | Needs understanding of probability distributions (Histogram/KDE)                 |
| **Prediction / Regression**          | Scatter Plots show trends (e.g., area vs. price), helpful in training models    |
| **Categorical Data Analysis**        | Count Plot, Bar Plot, Box Plot, Violin Plot                                      |
| **Distribution & Composition**       | Hist, KDE, Pie (if used), Count plots                                            |
| **Comparison**                       | Bar, Violin, Box, compare across categories                                     |
| **Relationship**                     | Scatter, Pair Plot, check feature correlations                                  |
