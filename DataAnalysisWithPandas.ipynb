ðŸ“Š Introduction to Statistics in Data Analysis

In data science, statistics helps summarize, describe, and draw conclusions from data.

ðŸ“‚ Types of Data
1. Quantitative Data
Numerical data (e.g., income, age, population).

2. Qualitative Data
Categorical or descriptive data (e.g., gender, city). Not present in this dataset.

3. Descriptive Statistics
Summarizes features of a dataset using:

Mean, Median, Mode

Range, Variance, Standard Deviation

4. Inferential Statistics

Uses sample data to make predictions or test hypotheses about a population.

5. Time Series Data

Data collected over time intervals (e.g., monthly sales).



ðŸ“Œ Measures of Central Tendency

These are used to identify the center or average of a data distribution.

| Measure    | Description              | Example     |
| ---------- | ------------------------ | ----------- |
| **Mean**   | Arithmetic average       | `.mean()`   |
| **Median** | Middle value when sorted | `.median()` |
| **Mode**   | Most frequent value      | `.mode()`   |


ðŸ”¹ Median (Odd Count): Middle value

ðŸ”¹ Median (Even Count): Average of two middle values

ðŸ“ Measures of Dispersion

Used to understand data spread and variability.

ðŸ”¸ 1. Range
Formula: max - min

ðŸ”¸ 2. Variance
Average of squared deviations from the mean

ðŸ”¸ 3. Standard Deviation (SD)
Square root of variance

Measures spread around the mean

ðŸ”¸ 4. Interquartile Range (IQR)
IQR = Q3 - Q1

Helps detect outliers

ðŸ“ˆ Quartiles and Quartile Deviation

| Quartile | Description            |
| -------- | ---------------------- |
| **Q1**   | 25% of data below this |
| **Q2**   | 50% (Median)           |
| **Q3**   | 75% of data below this |


Quartile Deviation (QD)


QD = (Q3 - Q1) / 2

Coefficient of QD

(Q3 - Q1) / (Q3 + Q1) â†’ A relative (unit-less) measure

ðŸ” Outlier Detection (Using IQR)

Lower bound: Q1 - 1.5 Ã— IQR

Upper bound: Q3 + 1.5 Ã— IQR

Any value outside this range is an outlier


| Measure Type | Description | Example                         |
| ------------ | ----------- | ------------------------------- |
| **Absolute** | Has units   | Range in dollars                |
| **Relative** | No units    | Ratio, Coefficient of Variation |
