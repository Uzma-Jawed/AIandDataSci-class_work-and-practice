```python
# ðŸ“Œ Visualizing Correlation Matrix using Seaborn

import seaborn as sns
import matplotlib.pyplot as plt

# Basic heatmap
sns.heatmap(data.corr())
plt.title("Correlation Heatmap")
```

```python
# Heatmap with values and color theme
sns.heatmap(data.corr(), annot=True, cmap='Blues')
plt.title("Correlation Heatmap - Blues")
```

```python
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap - Coolwarm")
```

```python
sns.heatmap(data.corr(), annot=True, cmap='viridis')
plt.title("Correlation Heatmap - Viridis")
```

```python
# ðŸ“Œ GroupBy and Aggregation

# Average house value for each age group
data.groupby("housing_median_age")['median_house_value'].mean().head()
```

```python
data.groupby("housing_median_age")['median_house_value'].mean()
```

```python
# Full summary of each age group
data.groupby("housing_median_age").mean().head()
```

```python
# Mean and Median of house values grouped by age
data.groupby('housing_median_age')['median_house_value'].agg(['mean', 'median']).head()
```

```python
# Aggregating multiple functions on multiple columns
data.aggregate({
    'housing_median_age': ['mean', 'sum'],
    'median_house_value': ['mean', 'max']
})
```

```python
# calculate mean etc using groupby of 3 columns
```

```python
# GroupBy with 3 Columns and Mean
grouped_mean = data.groupby(['housing_median_age', 'total_rooms', 'total_bedrooms'])['median_house_value'].mean()
grouped_mean.head()
```

```python
# GroupBy with Aggregate Functions
```

```python
grouped_stats = data.groupby(['housing_median_age', 'total_rooms', 'total_bedrooms'])['median_house_value'].agg(['mean', 'max', 'min', 'count'])
grouped_stats.head()
```

```python
# Reset Index to Flatten
```

```python
grouped_stats.reset_index().head()
```

## ðŸŸ© Variance:
A measure of how much the data points differ from the mean.

## ðŸŸ¦ Covariance:
Shows how two variables change together.
- Positive â†’ both increase together
- Negative â†’ one increases, the other decreases

## ðŸŸ¨ Correlation:
Normalized form of covariance, ranges from -1 to 1.
- +1 â†’ strong positive relationship
- -1 â†’ strong negative relationship
- 0 â†’ no relationship

## ðŸŸª groupby():
Used to group rows based on a column and apply summary stats like mean, count, etc.

## ðŸŸ¥ aggregate():
Used to apply multiple functions at once to grouped or entire data, useful in multi-metric analysis.
