3. Quartile Analysis & Outlier Detection

üìä Understanding Quartiles

Quartiles divide our data into four equal parts.


---



```python
# Calculate quartiles for age column
age_q1 = data['housing_median_age'].quantile(0.25)
age_q3 = data['housing_median_age'].quantile(0.75)
age_iqr = age_q3 - age_q1

print(f"Age Quartiles: Q1 = {age_q1}, Q3 = {age_q3}")
print(f"IQR: {age_iqr}")
```



---
üîç Quartile Deviation Metrics

These help us understand the spread of the middle 50% of our data.


---





```python
# Quartile Deviation (QD)
qd = age_iqr / 2
print(f"Quartile Deviation: {qd}")
```

```python
# Coefficient of QD
coeff_qd = (age_q3 - age_q1) / (age_q3 + age_q1)
print(f"Coefficient of QD: {coeff_qd:.4f}")
```



---

Interpretation:

A higher coefficient indicates greater dispersion

Values range from 0 to 1 (0 = no dispersion, 1 = maximum dispersion)

4. Comprehensive Outlier Analysis

üö® IQR Outlier Detection Method

We'll use the robust IQR method to identify unusual values in the 'households' column.


---



```python
column = 'households'
```

```python
# Calculate range
col_range = data[column].max() - data[column].min()
print(f"Range of {column}: {col_range}")
```

```python
# Calculate quartiles
q1 = data[column].quantile(0.25)
q3 = data[column].quantile(0.75)
iqr = q3 - q1
print(f"\nQuartiles: Q1 = {q1}, Q3 = {q3}")
```

```python
print(f"IQR: {iqr}")
```

```python
# Determine outlier boundaries
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
print(f"\nOutlier Thresholds:")
print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")
```

```python
# Identify outliers
outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
print(f"\nFound {len(outliers)} outliers ({len(outliers)/len(data)*100:.2f}% of data)")
```

```python
# Display outliers
outliers.head()
```



---

### Outlier Analysis Insights:

The IQR method identifies values outside 1.5√óIQR from the quartiles

**Outliers may represent:**

Data entry errors

Genuine extreme cases

Measurement anomalies

**Decisions needed on whether to:**

Remove outliers

Transform values

Keep them if meaningful
