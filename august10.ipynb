```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Sample Data
df = pd.DataFrame({
    'Weight': [70, 80, 60, 90],
    'Height': [1.75, 1.80, 1.65, 1.82],
    'City': ['Lahore', 'Karachi', 'Islamabad', 'Lahore']
})

# Numeric and categorical column lists
num_cols = ['Weight', 'Height']
cat_cols = ['City']

# Create ColumnTransformer
ct = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),                 # Scale numeric columns
        ('cat', OneHotEncoder(sparse_output=False), cat_cols) # One-hot encode categorical column
    ],
    remainder='passthrough'  # Correct spelling
)

# Fit and transform
result = ct.fit_transform(df)

# Get new column names
num_names = num_cols
cat_names = ct.named_transformers_['cat'].get_feature_names_out(cat_cols).tolist()
final_cols = num_names + cat_names

# Final DataFrame
df_transformed = pd.DataFrame(result, columns=final_cols)
print(df_transformed)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
# Create full pipeline: preprocessing + model
model_pipeline = Pipeline([
    ('preprocessing', ct),
    ('classifier', LogisticRegression())
])

# # Fit model
# model_pipeline.fit(X, y)

# # Predict
# preds = model_pipeline.predict(X)
# print("Predictions:", preds)
```
